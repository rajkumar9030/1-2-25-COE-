```python
#Median for Sales (to avoid outliers affecting the mean). 
median_data= data['Total Amount'].median()
data['Total Amount'].fillna(median_data, inplace=True)
data
```

```python
#Mode for Store and Item (most common value).
mode_data = data['Product Category'].mode()[0]
data['Product Category'].fillna(mode_data, inplace=True)
data
```

```python
#Drop rows with too many missing values.
limit = len(data) * 0.5 
data.dropna(thresh=limit, inplace=True)
```

```python
#Convert Date column to datetime format. 
#Extract new features from Date: Year, Month, Day, Weekday, Quarter.,Create a Weekend column (1 if Saturday/Sunday, else 0).
data['Date'] = pd.to_datetime(data['Date'], errors='coerce')
data['Year'] = data['Date'].dt.year
data['Month'] = data['Date'].dt.month
data['Day'] = data['Date'].dt.day
data['Weekday'] = data['Date'].dt.weekday
data['Quarter'] = data['Date'].dt.quarter
data['Weekend'] = data['Date'].dt.weekday >= 5
data
```

```python
# Removing Duplicates 
#Check for duplicate rows and remove them.
print("Duplicate rows before removal:", data.duplicated().sum())
data.drop_duplicates(inplace=True)
print("Duplicate rows after removal:", data.duplicated().sum())
```

```python
# Outlier Detection & Handling 
#Boxplots (to detect extreme values). 

import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x=data['Total Amount'])
plt.show()
```

```python
#Z-score or IQR method to remove/fix outliers. 
#Cap sales at a reasonable max limit.
from scipy.stats import zscore
z_scores = zscore(data['Total Amount'])
threshold = 3
data = data[(z_scores < threshold) & (z_scores > -threshold)]
print(f"Number of rows after Z-score outlier removal: {len(data)}")
```

```python
# One-Hot Encoding for 'Product Category'
data_encoded = pd.get_dummies(data, columns=['Product Category'], drop_first=True)
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['Product Category Encoded'] = label_encoder.fit_transform(data['Product Category'])
print(data['Product Category Encoded'])
```

```python
#Feature Engineering 
#Create New Features: Sales per Item = Sales / Item 
data['Sales per Item'] = data['Total Amount'] / data['Quantity']
```

```python
# To calculate sales growth, we need 'Total Amount' for previous days. 
# Assuming the data is sorted by 'Date'.
data['Date'] = pd.to_datetime(data['Date'])
data = data.sort_values(by='Date')

# Create the Sales Growth Rate feature
data['Sales Growth Rate'] = data['Total Amount'].pct_change()
```

```python
# Sales from previous day (lag of 1)
data['Sales Lag 1'] = data['Total Amount'].shift(1)

# Sales from previous week (lag of 7)
data['Sales Lag 7'] = data['Total Amount'].shift(7)
```

```python
# 7-day rolling average of sales
data['Rolling Avg 7'] = data['Total Amount'].rolling(window=7).mean()
```
